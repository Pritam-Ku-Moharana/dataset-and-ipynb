```python
# we have imbalanced dataset, so we have to do RandomOverSampling to adjust the data point in such a way so that the model doesnot get biased.
balanced_df = []

for grade, group in required_data.groupby("Grade"):
    if len(group) < 1000:
        # Oversample (duplicate)
        needed = 1000 - len(group)
        oversampled = group.sample(needed, replace=True)
        final_group = pd.concat([group, oversampled])
    else:
        # Undersample (take only 500)
        final_group = group.sample(1000, replace=False)
    balanced_df.append(final_group)

# Combine all
balanced_df = pd.concat(balanced_df).reset_index(drop=True)
# shape of balanced_df : 20500 rows  64 columns
```

```python
train_df , test_df = train_test_split(balanced_df,test_size=0.25, random_state=42)
```
